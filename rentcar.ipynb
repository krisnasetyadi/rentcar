```python
mobilmini=("daihatsu","avanza")
mobilsdn=("civic","bmw")
mobila=("daihatsu","avanza","civic","bmw")

def menu():
    print("menu sewa: ")
    print("1. mobil minibus")
    print("2. mobil sedan")
    print("3. pencarian")
    print("4. keluar")
username="admin"
password="admin"

username_input=input("masukan username: ")
password_input=input("masukan password: ")
if(username_input==username and password_input==password):
    menu()
else:
    username_input=input("masukan username: ")
    password_input=input("masukan password: ")
    while "admin" not in username_input and password_input:
        username_input=input("masukan username: ")
        password_input=input("masukan password: ")
        if("admin" in username_input and password_input):
            break



def menu2():
    print("menu mobil minibus: ")
    print(mobilmini)
def menu3():
    print("menu mobil sedan: ")
    print(mobilsdn)
def mobilminibus():
    print("mobil minibus")
    print("//per jam all flat 15.000")
    x= int(input("berapa lama sewa yang diinginkan = "))
    mobil_keluarga=x*15000
    print("total pembayaran: ",mobil_keluarga)
    ulang=str(input("mau mengulang sewa? [Y/N]"))
    if ulang=="Y" or "N":
        menu();
    else:
            exit()
def mobilsedan():
    print("mobil sedan")
    print("//per jam all flat 25.000")
    x= int(input("berapa lama sewa yang diinginkan = "))
    mobil_sport=x*25000
    print("total pembayaran: ",mobil_sport)
    ulang=str(input("mau mengulang sewa? [Y/N]"))
    if ulang=="Y" or "N":
        menu();
    else:
            exit()
def cari():
    cari = input("Masukkan nama mobil yang dicari: ")
    ketemu = False
    for i in range(0, len(mobila)):
        if cari == mobila[i]:
            ketemu = True
            break

    if ketemu:
        print("Mobil: ", cari, "berhasil ditemukan")
    else:
        print("Mobil: ", cari, "tidak ditemukan")
while 1:
    pilih=int(input("menu pilihan (ketik angka 4 jika ingin keluar): "))
    if pilih == 1:
        menu2()
        mobilminibus()
    elif pilih == 2:
        menu3()
        mobilsedan()
    elif pilih == 3:
        cari()
    elif pilih == 4:
        file_baca = open("test.txt", "r")
        print("\n")
        break;
```
